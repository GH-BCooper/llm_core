```python
from openai import OpenAI
from scraper import fetch_website_contents, fetch_website_links
from IPython.display import Markdown, display
```

```python
MODEL = "gemma4"
ollama = OpenAI(
    api_key="ollama",
    base_url="http://localhost:11434/v1"
)
```

```python
system_prompt = """You are an expert software engineer that  can summarize the the contents of a README file in the GitHub repository of a user.
You will be given a URL to a repository, and you will need to summarize the contents of the README file in the repository.
"""
```

```python
user_prompt = f"""I want you to summarize the contents of the README file in the repository at the given URL. 
Analyze the following GitHub README and provide:

# Project Summary
Write a short 2-3 sentence summary.

# Project Purpose
Explain what problem this project solves.

# Key Features
List the important features in bullet points.

# Tech Stack
Mention all technologies/frameworks/libraries used.

# Installation
Summarize the installation/setup process.

# Target Audience
Who would benefit from using this project?
"""
```

```python
def messages_for(website):
    return [
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": user_prompt + website}
    ]
```

```python
def summarize(url):
    website = fetch_website_contents(url)
    response = ollama.chat.completions.create(
        model = MODEL,
        messages = messages_for(website)
    )
    return response.choices[0].message.content
```

```python
def display_summary(url):
    summary = summarize(url)
    display(Markdown(summary))
```

```python
display_summary("https://github.com/GH-BCooper/ConnectNext-Real-Time-Chat/blob/main/README.md")
```
